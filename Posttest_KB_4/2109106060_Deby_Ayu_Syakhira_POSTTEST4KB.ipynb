4. __Melakukan Data cleaning pada Data dengan nilai null__

- __Mengecek apakah ada nilai null di dalam dataset__

```python
dataFrame.isna().sum()
```

- __Karena dataset tidak memiliki nilai null, maka langkah selanjutnya yaitu membuat dataset agar memiliki nilai null__

```python
kolom = ['Smoking', 'AlcoholDrinking', 'MentalHealth', 'SleepTime' ] #menentukan kolom yang akan diubah nilainya menjadi Null secara acak

probability = 0.5 #menentukan probabilitas munculnya nilai null setiap kolom, yaitu 50%

#menggunakan perulangan untuk menjelajahi setiap baris
for baris in range(len(dataFrame)):
    #memilih secara acak salah satu dari kolom yang sudah ditentukan sebelumnya
    kol = np.random.choice(kolom)
    #memeriksa apakah nilai acak kurang dari probabilitas yang telah ditentukan
    if np.random.rand() < probability:
        #jika ya, maka nilai di kolom tersebut diubah menjadi None (Null)
        dataFrame.at[baris, kol] = None

print(dataFrame) #menampilkan data yang sudah memiliki nilai null
print(dataFrame.isna().sum()) #menampilkan jumlah nilai null di setiap kolom
```

- __Mengganti nilai null sesuai dengan ketentuan__

```python
imputers = {} #membuat dictionary kosong untuk menyimpan imputers

#menggunakan perulangan untuk menjelajahi setiap kolom
for kol in kolom:
    #jika tipe data kolom adalah 'object', akan mengisi nilai yang Null dengan 'Unknown'
    if dataFrame[kol].dtype == 'object':
        dataFrame[kol].fillna('Unknown', inplace=True)
    #jika tipe data kolom adalah 'int64', akan mengisi nilai yang Null menggunakan SimpleImputer dengan strategi 'mean'
    elif dataFrame[kol].dtype == 'int64':
        imputer = SimpleImputer(strategy = 'mean')
        dataFrame[kol] = imputer.fit_transform(dataFrame[[kol]])
    #jika tipe data kolom adalah 'float64', akan mengisi nilai yang Null menggunakan SimpleImputer dengan strategi 'median'
    elif dataFrame[kol].dtype == 'float64':
        imputer = SimpleImputer(strategy = 'median')
        dataFrame[kol] = imputer.fit_transform(dataFrame[[kol]])

#menampilkan data setelah melakukan subtitusi nilai Null
print(dataFrame)
print(dataFrame.isna().sum())
```
