# Membaca data dari file CSV _heart_2020_cleaned_

```python
import pandas as pd                                         #mengimpor library pandas
import matplotlib.pyplot as plt
import seaborn as sns                                       #mengimpor library numpy
dataFrame = pd.read_csv('heart_2020_cleaned.csv')           #membaca file CSV dengan nama 'heart_2020_cleaned.csv' dan disimpan di dalam variabel dataFrame
dataFrame 
```

# 1. GRAFIK BATANG

```python
#menghitung jumlah data berdasarkan kategori usia dan mengurutkannya berdasarkan index
data = dataFrame['AgeCategory'].value_counts().sort_index()

#menghitung total jumlah data dalam dataset
total_data = len(dataFrame)

plt.figure(figsize = (12, 6)) #mengatur ukuran figure

#membuat grafik batang menggunakan Seaborn
gb = sns.barplot(x = data.index, y = data.values)

#membuat judul grafik dan label pada sumbu x dan y
plt.title('Distribusi Kategori Usia')
plt.xlabel('Kategori Usia')
plt.xticks(rotation = 45) #mengatur rotasi label sumbu x 
plt.ylabel('Jumlah')

#membuat label persentase setiap bar di grafik
#menggunakan perulangan untuk melalui setiap bar yang ada di grafik
for i in gb.patches:
    jdata = i.get_height() #mengambil tinggi (banyaknya) data dari setiap bar
    gb.annotate(f'{jdata/total_data:.2%}', (i.get_x() + i.get_width() / 2., jdata), ha = 'center', va = 'bottom')

plt.show()
```

# 2. GRAFIK PIE (PIE CHART)

```python
#menghitung jumlah data berdasarkan kolom 'HeartDisease'
jumlah = dataFrame['HeartDisease'].value_counts()

#menentukan jarak potongan 'Yes' dari pusat pie chart
explode = [0.2, 0] 

#membuat pie chart 
plt.pie(jumlah, labels = ['No', 'Yes'], colors = ['blue', 'gold'], autopct = '%1.1f%%', startangle = 180, explode = explode, shadow = True)

#membuat judul pie chart
plt.title('Jumlah Orang dengan dan tanpa Penyakit Jantung')

#membuat lebel legend di luar pie chart
labels = ['Tidak Punya Penyakit Jantung', 'Punya Penyakit Jantung']
plt.legend(labels, bbox_to_anchor = (1.5, 0.9))
plt.axis('equal') #memastikan  skala sumbu x dan y pada plot sama 


plt.show()
```

## Kesimpulan:
- Grafik pie di atas menggambarkan jumlah orang yang memiliki dan tidak memiliki penyakit jantung.
- Potongan "No" menunjukkan jumlah orang yang tidak memiliki penyakit jantung,
- Sedangkan potongan "Yes" menunjukkan jumlah orang yang memiliki penyakit jantung.
- Dari grafik di atas dapat dilihat sebanyak 91.4% orang yang tidak memiliki penyakit jantung dan 8.6% orang yang memiliki penyakit jantung.

# 3. HISTOGRAM

```python
#membuat variabel 'kolom' yang berisi daftar kolom yang akan digunakan untuk membuat histogram
kolom = ['BMI', 'PhysicalHealth', 'MentalHealth', 'SleepTime']

#membuat subplot dengan ukuran 2x2 dan mengatur ukuran figure
fig, gh = plt.subplots(2, 2, figsize = (12, 8))
fig.suptitle('Distribusi Data dengan Histogram: BMI, Physical Health, Mental Health, dan Sleep Time') #membuat judul figure

#menggunakan perulangan untuk membuat 4 subplot histogram, serta menambahkan garis rata-rata serta median
for i, column in enumerate(kolom):
    plt.subplot(2, 2, i + 1) #memilih subplot
    sns.histplot(dataFrame[column], bins=15, color = 'green') #membuat histogram dengan seaborn

    #menghitung nilai rata-rata dari data sesuai kolom dan menambahkan garis rata-rata pada histogram
    mean = dataFrame[column].mean()
    plt.axvline(mean, color = 'red', linestyle = 'solid', linewidth = 2, label = f'Rata-rata: {mean:.2f}')

    #menghitung nilai median dari data sesuai kolom dan menambahkan garis median pada histogram
    median = dataFrame[column].median()
    plt.axvline(median, color = 'yellow', linestyle = 'solid', linewidth = 2, label = f'Median: {median:.2f}')
    
    plt.title(f'{column}') #membuat judul subplot
    #membuat lebel pada sumbu x dan y
    plt.ylabel('Frekuensi kemunculan')
    plt.xlabel(column)
    #membuat lebel legend di luar histogram
    plt.legend()

plt.tight_layout() #mengatur tata letak subplot
plt.show()
```

## Kesimpulan:
- Grafik histogram di atas memvisualisasikan distribusi data dari empat kolom berbeda dalam satu gambar.
- Setiap subplot histogram menggambarkan distribusi data dari kolom yang berbeda, yaitu 'BMI', 'PhysicalHealth', 'MentalHealth', dan 'SleepTime'.
- Subplot pertama menggambarkan distribusi data dari kolom 'BMI'. Pada histogram ini, kita dapat melihat frekuensi kemunculan berbagai nilai BMI dari dataset. Rata-rata (mean) dari kolom 'BMI' ditandai dengan garis vertikal merah yang terletak pada nilai 28,33 pada sumbu x. Sementara itu, Nilai tengah (median) dari kolom 'BMI' ditandai dengan garis vertikal kuning yang terletak pada nilai 27,34 pada sumbu x.
- Subplot kedua menggambarkan distribusi data dari kolom 'PhysicalHealth'. Pada histogram ini, kita dapat melihat frekuensi kemunculan berbagai nilai PhysicalHealth dari dataset. Rata-rata (mean) dari kolom 'PhysicalHealth' ditandai dengan garis vertikal merah yang terletak pada nilai 3,37 pada sumbu x. Sementara itu, Nilai tengah (median) dari kolom 'PhysicalHealth' ditandai dengan garis vertikal kuning yang terletak pada nilai 0 pada sumbu x.
- Subplot ketiga menggambarkan distribusi data dari kolom 'MentalHealth'. Pada histogram ini, kita dapat melihat frekuensi kemunculan berbagai nilai MentalHealth dari dataset. Rata-rata (mean) dari kolom 'MentalHealth' ditandai dengan garis vertikal merah yang terletak pada nilai 3,90 pada sumbu x. # 5. LINE PLOT

```python
#membuat variable 'kel_usia' yang berisi data yang dikelompokkan berdasarkan 'AgeCategory',lalu menghitung rata-rata BMI untuk setiap kelompok usia
kel_usia = dataFrame.groupby('AgeCategory')['BMI'].mean()

plt.figure(figsize=(12, 6)) #mengatur ukuran figure

#membuat line plot dengan menggunakan data rata-rata BMI
plt.plot(kel_usia.index, kel_usia.values, marker='o', linestyle='-', color='gold', label='Rata-rata BMI')

#membuat judul grafik dan label pada sumbu x dan y
plt.title('Distribusi Rata-rata BMI berdasarkan Usia')
plt.xlabel('Usia (AgeCategory)')
plt.ylabel('Rata-rata BMI')

#memberikan nilai rata-rata BMI pada titik-titik di grafik
for i, value in enumerate(kel_usia.values):
    plt.text(kel_usia.index[i], value, f'{value:.2f}', ha='center', va='bottom', fontsize=10)

plt.grid(True) #menambahkan grid pada grafik
plt.legend() #membuat label legend pada grafik
plt.show()
```

# 6. HEATMAP

```python
#membuat variabel 'data' yang berisi daftar kolom yang akan digunakan untuk membuat heatmap
data = ['BMI', 'PhysicalHealth', 'MentalHealth', 'DiffWalking', 'AgeCategory', 'PhysicalActivity', 'SleepTime']
plt.figure(figsize=(5, 5)) #mengatur ukuran figure
matriks_corr = dataFrame[data].corr()#menghitung matriks korelasi antar kolom-kolom yang dipilih

#membuat heatmap
sns.heatmap(matriks_corr,  cmap='coolwarm', annot=True, linewidths=.5, center=0)

plt.title('Korelasi Antar Kolom')#membuat judul
plt.show()
```

## Kesimpulan:
- Grafik di atas memberikan gambaran visual mengenai hubungan antara kolom BMI, PhysicalHealth, MentalHealth, dan SleepTime dalam dataset.
  * Pada grafik, Korelasi antara BMI dan PhysicalHealth sebesar 0.11 artinya ada korelasi positif antara kedua kolom.
  * Pada grafik, Korelasi antara BMI dan MentalHealth sebesar 0.064 artinya ada korelasi positif antara kedua kolom.
  * Pada grafik, Korelasi antara BMI dan SleepTime sebesar -0.052 artinya ada korelasi negatif antara kedua kolom.
  * Pada grafik, Korelasi antara PhysicalHealth dan MentalHealth sebesar 0.29 artinya ada korelasi positif antara kedua kolom.
